深度優先廣度優先的實作
先把節點連接的節點先定義好

```
graph = {
    "A":["B","C"],
    "B":["A","C","D"], 
    "C":["A","B","D","E"],
    "D":["B","C","E","F"],
    "E":["C","D"],
    "F":["D"],      
}
def BFS(graph,s):
    queuelist = []
    queuelist.append(s)
    walk = set()
    walk.add(s)
    while(len(queuelist)>0):
        vertex = queuelist.pop(0)
        nodes = graph[vertex]
        for w in nodes:
            if w not in walk:
                queuelist.append(w)
                walk.add(w)
        print(vertex)
def DFS(graph,s):
    stacklist = []
    stacklist.append(s)
    walk = set()
    walk.add(s)
    while(len(stacklist)>0):
        vertex = stacklist.pop()
        nodes = graph[vertex]
        for w in nodes:
            if w not in walk:
                stacklist.append(w)
                walk.add(w)
        print(vertex)
print("BFS的順序")
BFS(graph,'A')
print("DFS的順序")
DFS(graph,'A')
```
