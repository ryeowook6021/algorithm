Selection sort (選擇排序)

```
def Selction_Sort(datas):
    for i in range(len(datas)-1): #n次
        min = i # 第一次選從第一個數開始,第二次從第二個數開始 #n次
        for j in range(i+1, len(datas)):    # 用記錄的max數與其他的比較 #n*n次
            if datas[j] < datas[min]: #n*n次
                min = j #n*n次

        datas[i], datas[min] = datas[min], datas[i] #n次

        print(datas)#n次
        print('---'*16)#n次
    return datas #1次
    #O(n^2) 


if __name__ == '__main__':
    import random 
    a = list(range(16))
    random.shuffle(a)
    print(a)
    a = Selction_Sort(a)
    print(a)
```
